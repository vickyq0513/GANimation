```
%%shell
for i in *.flv;
  do name=`echo "$i" | cut -d'.' -f1`
  echo "$name"
  ffmpeg -i "$i" -vf fps=1  /gdrive/My\ Drive/stanfordAI/"${name}%04d.jpg" -hide_banner
done
```
